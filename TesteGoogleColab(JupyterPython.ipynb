# Altmetrics 

````python
import requests
import time
import json
import csv

HEADERS = {
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.87 Safari/537.36',
}

ALTMETRIC_ENTRY = ["handle", "cited_by_fbwalls_count", "cited_by_feeds_count", "cited_by_gplus_count", "cited_by_msm_count", "cited_by_posts_count", "cited_by_rdts_count",

					"cited_by_tweeters_count", "cited_by_videos_count", "cited_by_accounts_count", "last_updated"]
# from https://docs.google.com/spreadsheets/d/1ndVY8Q2LOaZO_P_HDmSQulagjeUrS250mAL2N5V8GvY/edit#gid=0

#stats_file = open("stats_file.csv", "w", newline="")
stats_file = open("/content/altemetrics/stats_file.csv", "w")
writer = csv.writer(stats_file, delimiter=';', quotechar='"', quoting=csv.QUOTE_MINIMAL)

writer.writerow([entry for entry in ALTMETRIC_ENTRY])

handle_file = open("/content/altemetrics/handlesTesesDissertacoes_20200120_amostra10.csv", "r")
line_count = 0
for full_handle in handle_file:
  if line_count > 0:
	  handle_number = full_handle.split("http://hdl.handle.net/",1)[1]
	  api_str = "https://api.altmetric.com/v1/handle/" + handle_number
	  print(api_str)
	  response = requests.get(api_str.rstrip(), timeout=10, headers=HEADERS)
	  #OBS: em certos momentos o script cai com esse timeout=10, nao sei se nao e o servidor derrubando, pois quando retomo do ponto parado roda normal
	  if response.status_code == 404:
	  	time.sleep(1) #to do not hammer the server
	  	continue

	  json_response = json.loads(response.text)

	  csv_line = []

	  for entry in ALTMETRIC_ENTRY:
	  	if entry in json_response:
	  		if isinstance(json_response[entry], (float, int)):
	  			csv_line += [str(json_response[entry])]
	  		else:
	  			csv_line += [json_response[entry]]
	  	else:
	  		csv_line += "0"

	  writer.writerow([fields for fields in csv_line])

	  time.sleep(1) #to do not hammer the server
  line_count += 1
handle_file.close() 

stats_file.close()
````

````python
from IPython.display import Javascript

def target_func(comm, msg):
  # Only send the response if it's the data we are expecting.
  if msg['content']['data'] == 'the data':
    comm.send({
          'response': 'got comm open!',
        }, None, msg['buffers']);
get_ipython().kernel.comm_manager.register_target('comm_target', target_func)

Javascript('''
(async () => {
  const buffer = new Uint8Array(10);
  for (let i = 0; i < buffer.byteLength; ++i) {
    buffer[i] = i
  }
  const channel = await google.colab.kernel.comms.open('comm_target', 'the data', [buffer.buffer]);
  let success = false;
  for await (const message of channel.messages) {
    if (message.data.response == 'got comm open!') {
      const responseBuffer = new Uint8Array(message.buffers[0]);
      for (let i = 0; i < buffer.length; ++i) {
        if (responseBuffer[i] != buffer[i]) {
          console.error('comm buffer different at ' + i);
          return;
        }
      }
      // Close the channel once the expected message is received. This should
      // cause the messages iterator to complete and for the for-await loop to
      // end.
      channel.close();
    }
  }
  document.body.appendChild(document.createTextNode('done.'));
})()
''')
````
